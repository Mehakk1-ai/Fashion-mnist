```python
#CHANGE DIMENSION 3D TO 4D
#why ; Required for TensorFlow/Keras models expecting 4D input
#      Ensures compatibility with CNNs
train_images = np.expand_dims(train_images,-1)
test_images = np.expand_dims(test_images,-1)
```

convolutional neural network

```python
# Build CNN Model
model = tf.keras.Sequential([                     # creates a sequential model, meaning the layers are stacked in a linear order
    tf.keras.layers.Flatten(input_shape=(28,28)), # reshapes the input data from a 2D (28x28) array into a 1D (flattened) array
    tf.keras.layers.Dense(128,activation='relu'), #fully connected (dense) layer with 128 neurons
                                                  #ReLU helps introduce non-linearity and improves training efficiency.   
    tf.keras.layers.Dense(10)                     # output layer with 10 neurons.                   
])
```

```python
model.compile(optimizer='adam', #Adam (Adaptive Moment Estimation) is a widely used gradient-based optimization algorithm.
                                #Adam dynamically adjusts learning rates for each parameter, making training more efficient and stable.
                                #great default choice for deep learning models
              
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), #loss function, which measures the difference between predicted outputs and actual labels.
                                                                                      #from_logits=True tells TensorFlow that the output values are raw logits. 
              metrics =['accuracy'])          #
```

```python
model.fit(train_images,train_labels,epochs=10)
```

```python
test_loss,test_acc = model.evaluate(test_images,test_labels) #This function tests the model's performance using unseen data (test_images and test_labels).
                                                            #It calculates the loss and accuracy on the test set.
                                                             #test_loss → The computed loss on the test set.
                                                             #test_acc → The accuracy of the model on the test set.


print('\nTest accuracy' ,test_acc)
```

```python
model.fit (train_images , train_labels , epochs=10 , batch_size=512 , verbose=1  )

#model.fit() → This method starts the training process.
#epochs=10 ; Number of times the model goes through the entire training dataset
#batch_size=512 ;Number of samples processed before updating model weights.
#Controls the amount of information displayed during training:
          #0 → No output.
          #1 → Displays progress bar with loss/accuracy per epoch.
          #2 → Prints one line per epoch (less detailed than 1).
```
